```python
df_without_5percent_groups=df[df['application_type']=='INDIVIDUAL']
```

```python
valcounts=df_without_5percent_groups['application_type'].value_counts(normalize=True)*100
valcounts
```

```python
df=df_without_5percent_groups
null_counts = df.isnull().sum()
null_ratios = null_counts / df.shape[0]
columns_to_keep = null_ratios[null_ratios <= 0.2].index
df_filtered = df[columns_to_keep]
df.shape, df_filtered.shape, null_ratios[30:]
```

```python
num_unique = df.nunique()
columns_to_keep = num_unique[num_unique >= 10].index
df_filtered = df[columns_to_keep]
df.shape, df_filtered.shape 
removed_columns = set(df.columns) - set(df_filtered.columns)
print("Удаленные столбцы:", list(removed_columns))
```

```python
columns_to_drop = []
for col in df.columns:
    value_counts = df[col].value_counts(normalize=True)
    max_ratio = value_counts.max() if not value_counts.empty else 0
    if max_ratio > 0.8:
        columns_to_drop.append(col)
df = df.drop(columns=columns_to_drop)
df.shape
```

Выбросы присутствуют для процентной ставки и ежемесячного платежа

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

sns.boxplot(data=df, x='loan_status', y='int_rate', ax=axes[0,0])
axes[0,0].set_title('Процентные ставки по статусам займов')
axes[0,0].tick_params(axis='x', rotation=45)

sns.boxplot(data=df, x='loan_status', y='dti', ax=axes[0,1])
axes[0,1].set_title('Отношение долга к доходу по статусам займов')
axes[0,1].tick_params(axis='x', rotation=45)

sns.boxplot(data=df, x='loan_status', y='loan_amnt', ax=axes[1,0])
axes[1,0].set_title('Суммы кредитов по статусам займов')
axes[1,0].tick_params(axis='x', rotation=45)

sns.barplot(data=df, x='loan_status', y='annual_inc', ax=axes[1,1])
axes[1,1].set_title('Средний годовой доход по статусам займов')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
```

Проблемные займы имеют выше процентные ставки
Займы с высоким отношением долга к доходу чаще имеют проблемные статусы
Суммы кредитов распределены относительно равномерно по статусам
